```python
import nltk
from nltk import word_tokenize
```

```python
def printSomeIterable(someIter):
    for w in someIter:
        print(w)
```

```python
with open("CyberDocAsText.txt", "r") as f:
    contents = f.read()
```

```python
print(contents)
docWords = word_tokenize(contents)
vocab = set(docWords)
bg = nltk.bigrams(docWords)
```

```python
printSomeIterable(bg)
```
